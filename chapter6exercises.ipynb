```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=10000, noise=0.4, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42)

param_grid = {
    'max_leaf_nodes': list(range(2, 100)),
    'min_samples_split': [2, 4, 6, 8]
}

tree_cv = GridSearchCV(tree_clf, param_grid=param_grid, cv=3, verbose=2, scoring='accuracy')

tree_cv.fit(x_train, y_train)
```

```python
tree_cv.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

y_pred = tree_cv.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import ShuffleSplit

subsets = []

split_i = ShuffleSplit(n_splits=1000, test_size=len(x_train)-100, random_state=42)

for sub_train_i, sub_test_i in split_i.split(x_train):
    x_sub_train = x_train[sub_train_i]
    y_sub_train = y_train[sub_train_i]
    subsets.append((x_sub_train, y_sub_train))
```

```python
# b.
```

```python
from sklearn.base import clone
import numpy as np

# Clone ex.7 tree 1000 times to train on 1000 subsets
manual_forest = [clone(tree_cv.best_estimator_) for _ in range(1000)]

accuracy = []

for tree, (x_sub_train, y_sub_train) in zip(manual_forest, subsets):
    tree.fit(x_sub_train, y_sub_train)
    y_pred = tree.predict(x_test)
    accuracy.append(accuracy_score(y_test, y_pred))

np.mean(accuracy)
```

```python
from scipy.stats import mode

# Prediction matrix
y_pred = np.empty([1000, len(x_test)], dtype=np.uint8)

# Fill in predictions from trees
for tree_i, tree in enumerate(manual_forest):
    y_pred[tree_i] = tree.predict(x_test)
    
# Popularity vote
y_pred_fin, n_votes = mode(y_pred, axis=0)
```

```python
accuracy_score(y_test, y_pred_fin.reshape([-1]))
```
